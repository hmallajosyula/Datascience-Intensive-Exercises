```python
from dateutil.parser import parse

def date_function(array):
    for item in array:
     dt = parse(item)
     #print(dt)
     print(dt.strftime('%d/%m/%Y'))
```

```python
Some_dirty_dates = ['11/23/2015', '11-15-2015', '2015-12-24 06:23:00', 'Jan-15', '13-Jun-2015']
```

```python
Some_dirty_dates
```

```python
date_function(Some_dirty_dates)
```

Reading all text files in a directory into a Pandas df:

```python
import pandas as pd
import glob
```

```python
filelist = []
for files in glob.glob("*.txt"):
    filelist.append(files)
    
```

```python
df_list = [pd.read_table(file) for file in filelist]
big_df = pd.concat(df_list)
```

Convert to boolean

```python
x = ['1','0','1','0','0','1','1','0']
```

```python
def convert_bool(arr):
    bool_arr=[None] * len(arr)
    for i in range (0,len(arr)):
        arr[i]=int(arr[i])
        #print "The value to be converted into bool is %d" %arr[i]
        #print bool(arr[i])
        bool_arr[i] = bool(arr[i])
    return bool_arr
```

```python
convert_bool(x)
```
